+ Xác định cột không cần thiết, thực hiện loại bỏ để làm sạch.

```python
df.isnull().sum() # Kiểm tra giá trị thiếu trong DataFrame
df.drop(columns=['Last Update'], inplace=True) # Xóa cột 'Last Update' khỏi DataFrame
df
```

+ Chuyển đổi kiểu dữ liệu ở các cột cho phù hợp (ví dụ cột chứa dữ liệu ngày tháng)

```python
# Chuyển đổi cột Last Update thành ngày tháng năm
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
df['Month'] = df['ObservationDate'].dt.month
df['Day'] = df['ObservationDate'].dt.day
df['Year'] = df['ObservationDate'].dt.year
df.drop(columns=['ObservationDate'], inplace=True)
df
```

+ Tính tổng số ca nhiễm, tử vong, và khỏi bệnh trên toàn thế giới.

```python
print('Tổng số ca nhiễm: %d \nTử vong: %d \nHồi phục trên thế giới: %d.'%(df['Confirmed'].sum().astype(int), df['Deaths'].sum().astype(int), df['Recovered'].sum().astype(int)))
```

+ Tính tỷ lệ tử vong trên số ca nhiễm

```python
print('Tỷ lệ tử vong trên số ca nhiễm: %.2f%%'%(df['Deaths'].sum()/df['Confirmed'].sum()*100))
```

+ Lọc dữ liệu cho một quốc gia cụ thể

```python
print('Dữ liệu đã lọc theo quốc gia: Mainland China')
df_china = df[df['Country/Region'] == 'Mainland China']
df_china
```

+ Phân loại các quốc gia thành các nhóm dựa trên số lượng ca nhiễm

```python
bins = [0, 1000, 10000, 50000, 100000, float('inf')]
labels = ['<1k', '1k-10k', '10k-50k', '50k-100k', '>100k']
df['Case Group'] = pd.cut(df['Confirmed'], bins=bins, labels=labels)
df
```

Số ca nhiễm mới hàng ngày

Tính số ca nhiễm COVID-19 hàng ngày trung bình trong một tuần để đánh giá sự
biến động hàng tuần của dịch bệnh

```python
df_number_of_confirmed_daily = df.groupby('ObservationDate',dropna = True).agg(Total_Confirmed = ('Confirmed','sum')).reset_index()
df_number_of_confirmed_daily['Total_Confirmed'] = df_number_of_confirmed_daily['Total_Confirmed'].astype(int)
df_number_of_confirmed_daily
```

```python
df['Week'] = df['ObservationDate'].dt.isocalendar().week
df
```

```python
avg_confirmed_week = df.groupby('Week')['Confirmed'].mean().reset_index()
avg_confirmed_week
```

```python
df_covid_pivot = df[['ObservationDate', 'Country/Region', 'Confirmed', 'Deaths', 'Recovered']]
df_covid_pivot['Year'] = df_covid_pivot['ObservationDate'].dt.year
df_covid_pivot['Month'] = df_covid_pivot['ObservationDate'].dt.month
df_covid_pivot
```
